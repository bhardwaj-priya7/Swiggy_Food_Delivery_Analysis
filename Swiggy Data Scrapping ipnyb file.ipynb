```python
# sdtting the chrome to be automated and mentioning the URL of swiggy
base_url = "https://www.swiggy.com/"
driver = webdriver.Chrome()

# extracting the swiggy website
driver.get(base_url)
location_element = driver.find_element(by=By.XPATH, value='//*[@id="location"]')

# location = input("enter a city: ")
location = "Bangalore"
location_element.send_keys(location)
location_element.send_keys(Keys.ENTER)
time.sleep(3)

# clicking the total restaurent number by XPATH
driver.find_element(by=By.XPATH,
                    value='//*[@id="root"]/div[1]/div[1]/div/div[1]/div[1]/div/div[2]/div/div[2]/button[1]').click()
time.sleep(5)

# height = driver.get_window_size()
# height1 = height['height']
max_element_val = driver.find_element(by=By.XPATH,
                                      value='//*[@id="all_restaurants"]/div/div[1]/div/div/div/div[1]').text
# Extracting the maximum number of restaurents in the integer format
max_element_val = int(max_element_val.split()[0])

j = 0
k = 1000
# Running the while loop to extract the details of all restaurents by keeping the scrooling button at middile. 
while True:
    driver.execute_script(f"window.scrollTo({j}, {k});")
    time.sleep(1)
    ele = driver.find_elements(by=By.XPATH, value='//*[@id="all_restaurants"]/div/div[2]/div[1]/div/div')
    print(len(ele))
    if len(ele) >= (max_element_val // 4):
        break
    else:
        j = j + 500
        k = k + 500

print("we are here outside the while loop")

elements = driver.find_elements(by=By.XPATH, value='//*[@id="all_restaurants"]/div/div[2]/div[1]/div/div')
print(len(elements))
```

# Html List

```python
# Initialize an empty list
html_list = []

# Loop through the elements and get their innerHTML
for ele in elements:
    html = ele.get_attribute("innerHTML")
    # Append the innerHTML to the list
    html_list.append(html)

# Print the list of innerHTML
print(html_list)
```

```python
import pandas as pd

# Create a dictionary with the restaurant names
data = {'Restaurant Names': restaurant_names}

# Convert the dictionary into a dataframe
df10 = pd.DataFrame(data)

# Print the dataframe
print(df10)
```

```python
import re

# List to store the restaurant ratings
ratings = []

# Loop through the HTML strings and extract the ratings
for html in html_list:
    rating = re.findall('<span class="icon-star _537e4"></span><span>(.*?)</span>', html)
    if rating:
        ratings.append(rating[0])
    else:
        ratings.append(' ')
# Create a dictionary with the ratings
data = {'Rating': ratings}

# Convert the dictionary into a dataframe
df11 = pd.DataFrame(data)

# Print the dataframe
print(df11)
```

```python
import re
import pandas as pd

# List to store the prices for two
price_for_two = []

# Loop through the html strings and extract the prices for two
for html in html_list:
    price = re.findall('<div class="nVWSi">(.*?)</div>', html)
    if price:
        price_for_two.append(price[0])

# Create a dictionary with the prices for two
data = {'Price for Two': price_for_two}

# Convert the dictionary into a dataframe
df12 = pd.DataFrame(data)
df12

# Print the price column
#print(df12['Price for Two'])

```

```python
import re
import pandas as pd

# List to store the cuisines
cuisines = []

# Loop through the HTML strings and extract the cuisine information
for html in html_list:
    cuisine_list = re.findall('<div class="_1gURR" title="(.*?)">', html)
    cuisine = ", ".join(cuisine_list)
    cuisines.append(cuisine)

# Create a dictionary with the cuisines
data = {'Cuisines': cuisines}

# Convert the dictionary into a dataframe
df13 = pd.DataFrame(data)

# Print the dataframe
print(df13)
```

```python
import re
import pandas as pd

# List to store the restaurant data
restaurant_data = []

# Loop through the html strings and extract the restaurant data
for html in html_list:
#     restaurant_ids = re.findall('data-rest-id="(.*?)"', html)
    restaurant_links = re.findall('<a href="(.*?)">', html)
    name = re.findall('<div class="nA6kb">(.*?)</div>', html)
    rating = re.findall('<span class="icon-star _537e4"></span><span>(.*?)</span>', html)
    cuisines = re.findall('<div class="_1gURR" title="(.*?)">', html)
    price = re.findall('<div class="nVWSi">(.*?)</div>', html)
    delivery_time = re.findall('<div>(\d+) MINS</div>', html)
    
    # If price or cusins not found, set it to empty string
    if not price:
        price = [""]
    if not cuisines:
        cuisines = [""]
        
    # Add the restaurant data to the list
    restaurant_data.append([restaurant_links[0],name[0], rating[0], cuisines[0], price[0], delivery_time[0]])
    restaurant_data.append([restaurant_links[1],name[1], rating[1], cuisines[1], price[1], delivery_time[1]])
    restaurant_data.append([restaurant_links[2],name[2], rating[2], cuisines[2], price[2], delivery_time[2]])


# Create a dataframe from the restaurant data
df88 = pd.DataFrame(restaurant_data, columns=['Link','Restaurant Name', 'Rating', 'Cuisines', 'Price for Two', 'Delivery Time'])
df88
```
